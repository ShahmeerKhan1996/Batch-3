#### What are the top 10 average MSRP (Price) per car brand (rounded figure)? with Horizontal Bar Plot

```python
avg_msrp_by_make = df.groupby('Make')['MSRP'].mean().nlargest(10).round(0).sort_values(ascending=False)
avg_msrp_by_make
```

```python
avg_msrp_by_make.nlargest(10).plot(kind='barh', figsize=(10, 6), color='skyblue')
plt.title('Top 15 Car Brands by Average MSRP')
plt.xlabel('Average MSRP ($)')
plt.ylabel('Car Brand (Make)')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

#### How has the MSRP (price) changed over the years (rounded figure)? with Line Plot

```python
avg_msrp_by_year = df.groupby('Year')['MSRP'].mean().round(0)

print(avg_msrp_by_year)
```

```python
plt.figure(figsize=(10, 5))
avg_msrp_by_year.plot(kind='line', marker='o', color='teal')

plt.title('Average MSRP Over the Years')
plt.xlabel('Year')
plt.ylabel('Average MSRP ($)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

#### How Engine HP (power) and MPG (efficiency) have changed over the years? show with two line plots

Hint: Average of (city mpg + highway MPG)

```python
df['Average MPG'] = (df['city mpg'] + df['highway MPG']) / 2
yearly_stats = df.groupby('Year')[['Engine HP', 'Average MPG']].mean().round(1)
yearly_stats

```

```python
plt.figure(figsize=(10, 6))
plt.plot(yearly_stats.index, yearly_stats['Engine HP'], label='Average Horsepower', color='firebrick', marker='o')
plt.plot(yearly_stats.index, yearly_stats['Average MPG'], label='Average MPG', color='seagreen', marker='o')

plt.title('Average Horsepower and MPG Over the Years')
plt.xlabel('Year')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

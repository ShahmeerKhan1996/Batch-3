### Question 2:

How does time spent on social media compare to time spent on education among teens?

- Plot side-by-side histograms of Time on Social Media and Time on Education
- Find and compare the median time for both.
- Write simple comments to explain what the data shows

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 5))

sns.histplot(df['Time_on_Social_Media'], kde=True, bins=10, color='red', ax=axes[0])
axes[0].set_title('Time on Social Media')
axes[0].set_xlabel('Hours')
axes[0].set_ylabel('Count')

sns.histplot(df['Time_on_Education'], kde=True, bins=10, color='lightgreen', ax=axes[1])
axes[1].set_title('Time on Education')
axes[1].set_xlabel('Hours')
axes[1].set_ylabel('Count')

plt.tight_layout()
plt.show()

median_social = df['Time_on_Social_Media'].median()
median_education = df['Time_on_Education'].median()

print(f"Median time on Social Media: {median_social} hours")
print(f"Median time on Education: {median_education} hours")
```

### Comments:

Teens spend more time on social media than on educational activities.

The median time on social media is higher, and its distribution is more spread out.

Time spent on education is generally lower and more concentrated at smaller values.

This suggests teens prioritize social media over educational screen time.

### Question 3

- How can we detect and identify outliers in Daily Usage Hours, Exercise Hours, and Screen Time Before Bed among teens using the IQR method?
- Display the records corresponding to these outliers using boxplot.



```python
plt.figure(figsize=(10, 4))
sns.boxplot(x=df['Daily_Usage_Hours'], color='pink')
plt.title('Boxplot of Daily_Usage_Hours')
plt.show()

plt.figure(figsize=(10, 4))
sns.boxplot(x=df['Exercise_Hours'], color='pink')
plt.title('Boxplot of Exercise_Hours')
plt.show()

plt.figure(figsize=(10, 4))
sns.boxplot(x=df['Screen_Time_Before_Bed'], color='pink')
plt.title('Boxplot of Screen_Time_Before_Bed')
plt.show()

```

```python
Q1 = df['Daily_Usage_Hours'].quantile(0.25)
Q3 = df['Daily_Usage_Hours'].quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers_daily = df[(df['Daily_Usage_Hours'] < lower) | (df['Daily_Usage_Hours'] > upper)]
print("Outliers in Daily_Usage_Hours:")
print(outliers_daily[['Name', 'Daily_Usage_Hours']])
```

```python
Q1 = df['Exercise_Hours'].quantile(0.25)
Q3 = df['Exercise_Hours'].quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers_exercise = df[(df['Exercise_Hours'] < lower) | (df['Exercise_Hours'] > upper)]
print("Outliers in Exercise_Hours:")
print(outliers_exercise[['Name', 'Exercise_Hours']])
```

```python
Q1 = df['Screen_Time_Before_Bed'].quantile(0.25)
Q3 = df['Screen_Time_Before_Bed'].quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers_screen_time = df[(df['Screen_Time_Before_Bed'] < lower) | (df['Screen_Time_Before_Bed'] > upper)]
print("Outliers in Screen_Time_Before_Bed:")
print(outliers_screen_time[['Name', 'Screen_Time_Before_Bed']])
```

### Question 4

Analyze the correlation heatmap showing relationships between teen behaviors, screen usage, and mental health indicators.

Focus your attention on the following numeric variables:

- Daily_Usage_Hours, Sleep_Hours, Academic_Performance, Addiction_Level, Time_on_Social_Media,

- Time_on_Gaming, Time_on_Education, Apps_Used_Daily, Anxiety_Level, Depression_Level,

- Self_Esteem, Social_Interactions, Exercise_Hours, Screen_Time_Before_Bed,

- Phone_Checks_Per_Day, Family_Communication, Weekend_Usage_Hours


Identify two strong positive and two strong negative correlation involving any of the above variables.

Explain briefly what these correlations might suggest about teen behavior, wellbeing, or digital habits.

Use specific values from the heatmap to support your observation.



```python
numeric_cols = ['Daily_Usage_Hours', 'Sleep_Hours', 'Academic_Performance',
                'Addiction_Level', 'Time_on_Social_Media', 'Time_on_Gaming', 'Time_on_Education', 'Apps_Used_Daily',
                'Anxiety_Level','Depression_Level','Self_Esteem','Social_Interactions','Exercise_Hours','Screen_Time_Before_Bed',
               'Phone_Checks_Per_Day','Family_Communication','Weekend_Usage_Hours']

plt.figure(figsize=(15,10))
corr = df[numeric_cols].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

#### Two Strong Positive Correlations:

Daily_Usage_Hours ↔ Addiction_Level = 0.60

Teens who use their phones more frequently tend to have a higher addiction level. This is a strong and expected behavioral correlation.

Addiction_Level ↔ Apps_Used_Daily = 0.32

The more apps a teen uses daily, the more likely they are to show signs of phone addiction.

#### Two Strong Negative Correlations:


Addiction_Level ↔ Sleep_Hours = -0.22

Higher phone addiction is associated with reduced sleep, which may impact health and daily functioning.

Self_Esteem ↔ Depression_Level = -0.03

Teens with lower self-esteem may experience slightly higher levels of depression.

### Question 5

How does the average addiction level vary across different age groups? Show with line graph

```python
avg_addiction_by_age = df.groupby('Age')['Addiction_Level'].mean().reset_index()

plt.figure(figsize=(10,5))
sns.lineplot(data=avg_addiction_by_age, x='Age', y='Addiction_Level', marker='o', color='blue')
plt.title('Average Addiction Level by Age')
plt.xlabel('Age')
plt.ylabel('Average Addiction Level')
plt.grid(True)
plt.show()
```
